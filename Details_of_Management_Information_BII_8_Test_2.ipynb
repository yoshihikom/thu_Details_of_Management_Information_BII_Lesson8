## (2)統計的仮説検定
### 統計的仮説検定とは
標本を使って，母集団に関する統計的な"判断"を下す方法．判断の基準として，確率的な表現を使うことが特徴．単に検定とも呼ばれる．統計的仮説検定にはさまざまな種類があり，判断する対象も手法によってさまざま．統計的仮説検定は，推測統計学における実務への応用として，頻繁に使われる．

標本だけを使って「母集団を対象とした判断」を下すことは，本来とても難しい．そのため，検定の結果の解釈は慎重に行わなければならない．  
使用頻度がとても高く，自分たちの分析で使うこともあれば，他人の検定結果を見たうえで内容を解釈しなければならないことがある．  
統計的仮説検定を使うことの是非についてはこのあと議論するが，少なくとも検定の結果の解釈ができるようになっている必要性はあると言える．

### 1変量データの$t$検定
検定の手法にはさまざまありますが、まずは1変量データに対する$t$検定を説明します。

1変量データの$t$検定の特徴を簡単に述べます。 
- **対象**：平均値 
- **判断すること**：平均値が“ある値”と異なると言えるかどうか 

例えば、内容量が50gと書かれたスナック菓子があったとします。しかし、49gしか入っていないこともあれば51gのお菓子が入っていることもあるでしょう。すべての製品で完璧に50gというのは難しいものです。  
内容量には多少ばらつきがあるでしょうが、内容量の平均値は50gとなっていてほしいところではあります。  
こんな時には検定の出番です。1変量データの$t$検定を使えば「内容量の平均値が50gと異なっていると言えるかどうか」を判断することができます。

$t$検定は1変量データ以外にも適用可能ですが、この章では1変量データのみを対象とします。そのため、単に$t$検定と呼称します。

### 有意差 
統計的仮説検定においては、しばしば有意差という言葉が使われます。文字通りの意味でとらえると「意味の有る差」となるでしょうか。  
意味の有る差という言葉があるということは、意味の無い差もあるということです。検定という枠組みにかかわらず、意味の有る差・意味の無い差について考えをめぐらすことは、データを解釈するにあたってとても重要なことといえます。

### $t$検定：直観的な考え方 

$t$検定における有意差の直観的な考え方を説明します。 

例えば、2袋のスナック菓子を開けたとしましょう。  
家にある安物の重量計でスナック菓子の内容量を測りました。1袋目が55gで、2袋目が44g、平均内容量が49.5gだったとしましょう。50gとは異なっていますが、これは「意味の有る差」といえるでしょうか。  
まず気になるのは2袋しか開けていないことです。もっと大きな標本を使って調査すべきです。  
さらに、測定方法も気になります。安物の重量計で正確な重量がわかるものでしょうか。  
また49.5gという結果は、50gとかなり近いです。  
この調査結果からは、有意差があると主張するのは難しそうに感じます。 

逆に、以下の条件であれば50gとの“意味の有る差”すなわち有意差があるように感じられます。 

 
- **大きなサンプルで調査した：サンプルサイズが大きい** 
- **精密な重量計で測定した：データのばらつき(分散)が小さい** 
- **重量の平均値が50gから大きく離れている：平均値の差が大きい** 

$t$検定では、この3つの条件が満たされている時に、有意差ありと判断します。

### 平均値の差が大きいだけでは有意差は得られない 

$t$検定の有意差判断ロジックからは、とても重要な教訓が得られます。それは、**平均値の差が大きいだけでは有意差が得られない**ということです。  
これはとても重要なことなので、ぜひ覚えておいてください．  
とても極端な例を挙げて、これを説明します。  
古い重量計があります。これの中に入っているバネはゆがんでいて、まったく同じ物の重量を測ったとしても、値が毎回変わるというひどい代物です。50gちょうどの重石の重量を測っても、30gになったり70gになったりします。この重量計を使って、2袋の重量を測りました。  
1つ目が10gで2つ目が60gでした。平均は35gであり、50gからはかなり離れているように見えます。  
しかし、この結果を見て有意差ありと考える人はまずいないでしょう。この重量計の測定値は、ばらつき（分散）があまりにも大きすぎ、サンプルサイズはか細いほどに小さいのですから。  

たまに平均値だけを使ってその大小を比較している記事を見かけますが、これは問題のある比較だということがわかるでしょう。サンプルサイズとデータのばらつき（分散）の大きさを加味することが必要なのです。  
統計的仮説検定という枠組みにかかわらず、意味の有る差について考えることは、データを読むリテラシーとして重要かと思います。

### $t$値 
$t$検定における、有意差ありと考える条件は以下の3つでした。 

- **大きなサンプルで調査した：サンプルサイズが大きい** 
- **精密な重量計で測定した：データのばらつき（分散）が小さい** 
- **重量の平均値が50gから大きく離れている：平均値の差が大きい** 

この3つをすべて合わせた指標を$t$値と呼び、以下のように計算されます。

$\displaystyle t値=\frac{標本平均-比較対象値}{標準偏差\div\sqrt{サンプルサイズ}}=\frac{標本平均-比較対象値}{標準誤差}$

スナック菓子の例を使って、50gと異なるかどうかの検定を行う場合は以下のようになります。

$\displaystyle t値=\frac{標本平均-50}{標準誤差}$

この時$t$値が大きければ50gと「有意差あり」と判断ができることになります。  
なお、標本平均が比較対象と比べてとても小さかった場合は，$t$値が小さな値になります。$t$値はその絶対値に意味が有ると思うとよいでしょう。

### 統計的仮説検定の枠組み：帰無仮説・対立仮説 
統計的仮説検定では、ある仮説を立てて、その仮説を棄却するかしないかという判断を下すことで、データからの客観的な判断を試みます。

棄却される対象となる最初の仮説を**帰無仮説**と呼びます。  
帰無仮説と対立する仮説を**対立仮説**と呼びます。

スナック菓子の平均値が50gといえるかどうかを判断する場合には、以下のように仮説を設定します。  
帰無仮説：スナック菓子の平均重量は50gである  
対立仮説：スナック菓子の平均重量は50gと異なる

帰無仮説が棄却された、すなわち帰無仮説が間違っていると判断されたならば、有意差あり、すなわち「スナック菓子の平均重量は50gと異なる」と判断することになります。  
少々遠回りして判断を下すことになりますが、厳密性を重んじた結果です。

### $p$値 
$p$値とは、標本と帰無仮説との矛盾の目安となる指標です。$p$値が小さいほど、帰無仮説と標本が矛盾していると考えます。  
$p$値は確率として表現されます。  
$p$値も信頼区間と同様に、「まったく同じ条件で何度も何度も標本抽出～$t$値の計算を繰り返す」ということをやって、確率を解釈できることに注意してください。この解釈は誤解を生みやすいため、今回の最後の(5)でも再度取り上げます。

### 有意水準
$p$値が有意水準を下回った時に，帰無仮説を棄却する．  
**有意水準**とは，帰無仮説を棄却する基準となる値だと思えばよい．有意水準としては$5\%$が使われることが多く，ここでもそれにならう．有意水準は**危険率**ともよばれる．

### $t$検定と$t$分布の関係
$t$値の計算式を再掲する．  

$\displaystyle t値=\frac{標本平均-50}{標準偏差}$

ここで，母平均を50と仮定すると，この式は以前に特徴を説明した$t$値と一致していることがわかります．  
ということは，母集団分布が平均50の正規分布であると仮定すると，標本の$t$値は$t$分布に従うことになります．

今回の標本から計算されてt値の絶対値を$t_{標本}$と呼ぶことにする．  
$t$分布の累積分布関数を使うと「母平均を50と仮定したときに，$t$値が$t_{標本}$を下回る確率」を計算することができる．この確率を$a$とします．  
$1-a$を求めると「母平均が50と仮定したときに，$t$値が$t_{標本}$を上回る確率」が計算できることになる．  
$1-a$が小さければ「$t$値が$t_{標本}$を上回る確率が低い（すなわち，$t_{標本}$は十分に大きい）」ということになり，有意差がえられそうです．

### 片側検定・両側検定
**片側検定**とは「スナック菓子の平均重量は$50g$よりも小さい」といったことを調べる検定方法です．$50g$よりも大きいと言うことは想定しません．  
片側検定の逆パターンとして「スナック菓子の平均重量は$50g$よりも大きい」ことだけを調べて，50gより小さいことは一切想定しないというものもあります．

**両側検定**は「スナック菓子の平均重量は$50g$と異なる」ということを調べる検定手法です．$50g$よりも小さいかもしれないし大きいかもしれません．

$p$値の計算の際に，片側検定か両側検定かの違いが出てきます．  
片方だけしか想定しないのは不自然ですので，両側検定を行うことが多いです．

### $p$値の計算方法
両側検定を行うことを前提として$p$値を計算する．

今回の標本から計算された$t$値のことを$t_{標本}$と呼ぶことにする．  
$t$分布の累積分布関数を使うと「母平均を50と仮定したときに，$t$値が$t_{標本}$を下回る確率」を計算することができる．この確率を$a$とする．

$p$値は以下のように計算される．

$p値=(1-a)\times2$

最後に$\times2$をしているのは両側検定だからです．「スナック菓子の平均重量は$50g$と異なる」確率を計算しようと思ったら，大きい場合と小さい場合の２つを考慮するため，$1-a$を2倍する必要がある．．  
なお，片側検定の場合は，単に$1-a$が$p$値となる．

### $t$検定の実装：分析の準備
まずは，必要なライブラリの読込や表示桁数の設定などを行います．

```python
# 数値計算に使うライブラリ
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

# グラフを描画するライブラリ
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

# 表示桁数の指定
%precision 3
# グラフをjupyter Notebook内に表示させるための指定
%matplotlib inline
```

続いて，今回の分析の対象となるデータを読み込みます．スナック菓子の重量を測定した架空のデータです．サンプルサイズは20です．シリーズ形式で読み込みます．

```python
# データの読み込み
junk_food = pd.read_csv(
    "ds2-testing-2-junk-food-weight.csv")["weight"]
junk_food.head()
```

```python
junk_food
```

このデータを対象に，1標本の$t$検定を実行します．  
以下の要領で検定を行います．  
帰無仮説：スナック菓子の平均重量は$50g$である  
対立仮説：スナック菓子の平均重量は$50g$と異なる  
有意水準は$5\%$とします．$p$値が0.05を下回れば，帰無仮説は棄却され，スナック菓子の重量は有意に$50g$と異なると主張ができることになります．

### $t$検定の実装：$t$値の計算
まずは，$t$値を計算します．  
$t$値の計算式を再掲します．比較対象は$50g$であることに注意します．

$\displaystyle t値=\frac{標本平均-比較対象値}{標準偏差}$

まずは標本平均を求めます．

```python
# 標本平均
mu = np.mean(junk_food)
mu
```

続いて自由度です．サンプルサイズから１を引きます．

```python
# 自由度
df = len(junk_food) - 1
df
```

次に標準誤差を求めます．標準誤差は「標準偏差$\div$サンプルサイズの平方根」で求められます．

```python
# 標準誤差
sigma = np.std(junk_food, ddof = 1)
se = sigma / np.lib.scimath.sqrt(len(junk_food))
se
```

最後に$t$値を計算します．

```python
# t値
t_value = (mu - 50) / se
t_value
```

### $t$検定の実装：$p$値の計算
続いて$p$値の計算に移ります．  
母集団に正規分布を仮定すると，$t$値は$t$分布に従うと考えられるので，$t$分布の累積分布関数が使えます．

繰り返しになりますが$p$値の計算方法をおさらいします．  
$t$分布の累積分布関数を使うと「母平均を50と仮定したときに，$t$値が$t_{標本}$を下回る確率」を計算することができます．この確率を$a$とします．  
$1-a$を求めると「母平均が50と仮定したときに，$t$値が$t_{標本}$を上回る確率」が計算できることになる．  
$1-a$が小さければ「$t$値が$t_{標本}$を上回る確率が低い（すなわち，$t_{標本}$は十分に大きい）」ということになり，有意差がえられそうです．  
両側検定を実施するために，$p$値は$(1-a)\times2$として計算されます．

```python
# p値
alpha = stats.t.cdf(t_value, df = df)
(1 - alpha) * 2
```

$p$値が有意水準0.05を下回っているので，有意差ありとみなせます．スナック菓子の平均重量は$50g$と有意に異なっていると判断できます．

なお，`stats.ttest_1samp`関数を使うと，もっと簡単に1標本の$t$検定を行うことができます（出力結果の桁数は少し短くさせてます）．

```python
# t検定
stats.ttest_1samp(junk_food, 50)
```

`statistic`が$t$値であり，`pvalue`が$p$値です．

### シミュレーションによるp値の計算
$p$値の意味を解釈するために，$p$値をシミュレーションで求めてみる．

$p$値は「帰無仮説が正しいと仮定して，何度も標本抽出～$t$値計算を繰り返したとき，$t_{標本}$と同じかそれより大きな$t$値が得られる割合」と解釈できる．両側検定の場合はこの割合を2倍したものが$p$値となる．  
この割合が小さいということは，$t_{標本}$を超えることがめったにないということです．$t_{標本}$が十分に大きいと考えられ，有意差が得られると判断できます．

まずは，今回の標本の情報を（一部再掲となるが）変数に格納しておきます．サンプルサイズと標準偏差を格納しました．

```python
# 標本の情報(一部再掲)
size = len(junk_food)
sigma = np.std(junk_food, ddof = 1)
```

シミュレーションをして50,000回$t$値を計算します．そこで，50,000個の$t$値を格納する入れ物を用意します．

```python
# t値を格納する変数
t_value_array = np.zeros(50000)
```

`stats.norm`に指定する平均値を50とすることで帰無仮説が正しいと仮定して，50,000回，標本抽出～$t$値計算を繰り返します．

```python
# 母平均が50(帰無仮説が正しい)と仮定してt値を計算することを50000回繰り返す
np.random.seed(1)
norm_dist = stats.norm(loc = 50, scale = sigma)
for i in range(0, 50000):
    sample = norm_dist.rvs(size = size)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof = 1)
    sample_se = sample_std / np.lib.scimath.sqrt(size)
    t_value_array[i] = (sample_mean - 50) / sample_se
```

50,000個の$t$値のうち，$t_{標本}$を上回った割合を求めます．これに２をかけると$p$値となります．

```python
(sum(t_value_array > t_value) / 50000) * 2
```

理論上の値とほぼ一致していることが確認できる．
